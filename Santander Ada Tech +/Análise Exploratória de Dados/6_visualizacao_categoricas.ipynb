### Importando as bibliotecas

```python
# Importa a biblioteca NumPy, que fornece suporte para arrays multidimensionais e funções matemáticas avançadas
# NumPy é amplamente utilizado em ciência de dados para manipulação eficiente de dados numéricos e cálculo matemático
import numpy as np

# Importa o Seaborn, uma biblioteca para visualização de dados baseada no Matplotlib
# O Seaborn facilita a criação de gráficos mais sofisticados e esteticamente agradáveis com menos código
import seaborn as sns

# Importa o Matplotlib, uma biblioteca para criação de gráficos e visualizações em Python
# O Matplotlib fornece controle total sobre o layout dos gráficos e é usado em conjunto com o Seaborn
import matplotlib.pyplot as plt
```

```python
# Carrega o conjunto de dados 'iris' utilizando a função 'load_dataset' do Seaborn
# O dataset 'iris' contém informações sobre flores da espécie Iris, com colunas como:
# - 'sepal_length': comprimento da sépala
# - 'sepal_width': largura da sépala
# - 'petal_length': comprimento da pétala
# - 'petal_width': largura da pétala
# - 'species': tipo da espécie da flor (Setosa, Versicolor ou Virginica)
iris = sns.load_dataset('iris')

# Exibe o DataFrame 'iris' contendo os dados carregados
# O comando 'iris' no final exibe o DataFrame no Jupyter Notebook ou no console
iris
```

```python
# Conta o número de ocorrências de cada valor único na coluna 'species' do DataFrame 'iris'
# A função 'value_counts()' retorna a quantidade de vezes que cada categoria (espécie de flor) aparece na coluna 'species'
# Neste caso, as três espécies de flores ('setosa', 'versicolor' e 'virginica') são contadas
iris['species'].value_counts()
```

```python
# A variável 'col' é atribuída ao nome da coluna 'sepal_length' que será usada para o gráfico
# Esta variável é útil para exibir dinamicamente o nome da coluna no título do gráfico
col = 'sepal_length'  # Corrigindo a grafia da variável

# A função 'sns.histplot' do Seaborn é usada para criar um histograma da coluna 'sepal_length' do DataFrame 'iris'
# O parâmetro 'x' especifica a coluna cujos valores serão plotados no eixo X do gráfico
# O argumento 'kde=True' adiciona uma linha de densidade (Kernel Density Estimate) sobre o histograma,
# mostrando a distribuição estimada dos dados de forma suave.
# O método 'set_title' adiciona um título ao gráfico, utilizando a variável 'col' para exibir o nome da coluna
sns.histplot(data=iris, x='sepal_length', kde=True).set_title(f'Distribuição da variável {col}')
```

```python
# Itera sobre todas as colunas do DataFrame 'iris' e imprime o nome de cada coluna
# O 'iris' é um DataFrame do Pandas e o laço 'for' percorre cada nome de coluna (por padrão, o nome das colunas são strings)
for col in iris:
    print(col)
```

```python
# Itera sobre todas as colunas do DataFrame 'iris', exceto a coluna 'species'
# O método 'drop(columns="species")' remove temporariamente a coluna 'species' do DataFrame para não ser considerada na iteração
for col in iris.drop(columns='species'):
    
    # Cria um histograma com uma linha de densidade (KDE) para cada coluna do DataFrame 'iris'
    # O parâmetro 'data=iris' indica que o DataFrame 'iris' está sendo usado para plotar os dados
    # O parâmetro 'x=col' diz ao Seaborn para usar cada coluna (exceto 'species') no eixo X
    # O argumento 'kde=True' adiciona a linha de densidade ao histograma
    sns.histplot(data=iris, x=col, kde=True).set_title(f'Distribuição da variável {col}')
    
    # Exibe o gráfico gerado após a criação de cada histograma
    plt.show()
```

```python
# Itera sobre todas as colunas do DataFrame 'iris', exceto a coluna 'species'
# O método 'drop(columns="species")' remove temporariamente a coluna 'species' do DataFrame para não ser considerada na iteração
for col in iris.drop(columns='species'):
    
    # Cria um histograma com uma linha de densidade (KDE) para cada coluna do DataFrame 'iris'
    # O parâmetro 'data=iris' indica que o DataFrame 'iris' está sendo usado para plotar os dados
    # O parâmetro 'x=col' diz ao Seaborn para usar cada coluna (exceto 'species') no eixo X
    # O argumento 'kde=True' adiciona a linha de densidade ao histograma
    # O argumento 'hue=iris['species']' permite dividir a visualização por categoria da coluna 'species',
    # colorindo os histogramas de acordo com as diferentes espécies de flores
    sns.histplot(data=iris, x=col, kde=True, hue=iris['species']).set_title(f'Distribuição da variável {col}')
    
    # Exibe o gráfico gerado após a criação de cada histograma
    plt.show()
```

```python
# Itera sobre todas as colunas do DataFrame 'iris', exceto a coluna 'species'
# O método 'drop(columns="species")' remove temporariamente a coluna 'species' do DataFrame para não ser considerada na iteração
for col in iris.drop(columns='species'):
    
    # Cria um gráfico de barras para cada coluna numérica do DataFrame 'iris'
    # O eixo X representa as diferentes espécies de flores
    # O eixo Y mostra os valores da variável 'col' (uma das colunas numéricas)
    # 'hue=iris['species']' faz com que as barras sejam coloridas de acordo com as espécies
    # 'ci=90' define o intervalo de confiança de 90% para as barras do gráfico
    sns.barplot(data=iris, x=iris['species'], hue=iris['species'], y=col, errorbar=('ci', 90))
    
    # Exibe o gráfico gerado após a criação de cada histograma
    plt.show()
```
